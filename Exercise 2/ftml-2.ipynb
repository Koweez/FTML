```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
np.random.seed(0)
n_samples = 1000  # Number of samples
lambda_param = 1  # Rate parameter for exponential distribution

# Generate data
X = np.random.normal(0, 1, n_samples)  # Generating X from a normal distribution (mean=0, std=1)
Y_squared = X + np.random.normal(0, 0.5, n_samples)  # Generating Y from a normal distribution (mean=0, std=0.5)
Y_absolute = np.abs(Y_squared)  # Creating absolute loss function

# Compute empirical test errors
error_squared = np.mean((Y_squared - np.mean(Y_squared))**2)  # Squared loss
error_absolute = np.mean(np.abs(Y_absolute - np.median(Y_absolute)))  # Absolute loss

print(f"Empirical test error (squared loss): {error_squared:.4f}")
print(f"Empirical test error (absolute loss): {error_absolute:.4f}")

# Plotting the results
plt.figure(figsize=(8, 6))
plt.hist(Y_squared, bins=30, alpha=0.5, label='Y (Squared Loss)', color='blue')
plt.hist(Y_absolute, bins=30, alpha=0.5, label='Y (Absolute Loss)', color='green')
plt.axvline(np.mean(Y_squared), color='blue', linestyle='dashed', linewidth=1, label='Mean (Squared Loss)')
plt.axvline(np.median(Y_absolute), color='green', linestyle='dashed', linewidth=1, label='Median (Absolute Loss)')
plt.title('Histogram of Y with Squared and Absolute Loss')
plt.xlabel('Y')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```
