```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
mu_X = 2000
sigma_X = 500
alpha = 0.1
beta = 50
sigma_Y = 20

# Proposed estimator parameters
alpha_tilde = 0.08
beta_tilde = 60

# Generate dataset
n_samples = 10000
X = np.random.normal(mu_X, sigma_X, n_samples)
Y = np.random.normal(alpha * X + beta, sigma_Y, n_samples)

# Define estimators
def f_star(x):
    return alpha * x + beta

def f_tilde(x):
    return alpha_tilde * x + beta_tilde

# Compute empirical risks
def empirical_risk(f):
    predictions = f(X)
    return np.mean((Y - predictions) ** 2)

bayes_risk_empirical = empirical_risk(f_star)
proposed_risk_empirical = empirical_risk(f_tilde)

# Results
bayes_risk_empirical, proposed_risk_empirical
```
