```python
import numpy as np

# Given parameters
n = 100  # number of samples
d = 5    # number of features
sigma_true = 2  # true standard deviation of the noise

# Simulate data
np.random.seed(0)
X = np.random.randn(n, d)
theta_star = np.random.randn(d)
epsilon = sigma_true * np.random.randn(n)
y = X @ theta_star + epsilon

# Estimate theta using OLS
theta_hat = np.linalg.inv(X.T @ X) @ X.T @ y

# Calculate residuals
residuals = y - X @ theta_hat

# Estimate sigma^2 using the result from Question 6
sigma_est = np.sum(residuals**2) / (n - d)

# Check consistency with the theoretical value
sigma_est, sigma_true**2
```
